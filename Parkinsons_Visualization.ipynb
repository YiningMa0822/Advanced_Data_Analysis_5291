```python
# Create a copy of the original X DataFrame to add 'period_id' without altering the original
X_copy = X.copy()

# Initialize 'period_id' column in the copy with zeros
X_copy['period_id'] = 0

# Set the initial period_id value in the copy
X_copy.loc[0, 'period_id'] = 1

# Loop through the DataFrame copy from the second row
for i in range(1, len(X_copy)):
    # If 'test_time' increased or remained the same, assign the same period_id as the previous row in the copy
    if X_copy.loc[i, 'test_time'] >= X_copy.loc[i-1, 'test_time']:
        X_copy.loc[i, 'period_id'] = X_copy.loc[i-1, 'period_id']
    # If 'test_time' decreased, increment the id by 1 in the copy
    else:
        X_copy.loc[i, 'period_id'] = X_copy.loc[i-1, 'period_id'] + 1

# Move the 'period_id' column to the first position in the copy
X_with_period_id = X_copy[['period_id'] + [col for col in X_copy.columns if col != 'period_id']]

# Display the first few rows of the new DataFrame to verify the 'period_id' column addition
X_with_period_id.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming DataFrames X, y, and X_with_period_id are already loaded in the environment
# Concatenating the DataFrames X and y to get 'total_UPDRS' in one DataFrame
merged_df = pd.concat([X, y], axis=1)

# Merging with X_with_period_id to get 'period_id'
merged_df_with_period = pd.merge(X_with_period_id[['period_id']], merged_df, left_index=True, right_index=True)

# Function to calculate the slope between two points
def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1) if x2 != x1 else np.inf

# Plotting the final graph
plt.figure(figsize=(10, 6))

# Extracting unique period IDs
period_ids_with_period = merged_df_with_period['period_id'].unique()

for period_id in period_ids_with_period:
    period_data_with_period = merged_df_with_period[merged_df_with_period['period_id'] == period_id].sort_values('test_time')

    # Plotting the lines with 20% opacity
    plt.plot(period_data_with_period['test_time'], period_data_with_period['total_UPDRS'], color='orange', alpha=0.2)

    # Marking the left and right endpoints with circle markers in dark red with 50% opacity
    plt.scatter(period_data_with_period['test_time'].iloc[[0, -1]],
                period_data_with_period['total_UPDRS'].iloc[[0, -1]],
                color='darkred', alpha=0.5, marker='o')

    # Checking for slope change points with 'x' markers in dark red with 50% opacity
    for i in range(1, len(period_data_with_period) - 1):
        if 75 < period_data_with_period['test_time'].iloc[i] < 125:
            prev_slope = calculate_slope(period_data_with_period['test_time'].iloc[i - 1],
                                         period_data_with_period['total_UPDRS'].iloc[i - 1],
                                         period_data_with_period['test_time'].iloc[i],
                                         period_data_with_period['total_UPDRS'].iloc[i])

            next_slope = calculate_slope(period_data_with_period['test_time'].iloc[i],
                                         period_data_with_period['total_UPDRS'].iloc[i],
                                         period_data_with_period['test_time'].iloc[i + 1],
                                         period_data_with_period['total_UPDRS'].iloc[i + 1])

            if abs(next_slope - prev_slope) > 0.01:
                plt.scatter(period_data_with_period['test_time'].iloc[i],
                            period_data_with_period['total_UPDRS'].iloc[i],
                            color='darkred', alpha=0.5, marker='x')

plt.title('Total UPDRS vs Test Time with UPDRS Recording Time Points (0, 3rd, and 6th month)')
plt.xlabel('Test Time')
plt.ylabel('Total UPDRS')
plt.grid(True)
plt.ylim(bottom=0, top=70)  # Set the upper limit to 70

plt.show()
```
